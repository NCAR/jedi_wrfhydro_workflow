```python

def preprocess_owp_obs_swe(ds):
    path = ds.encoding['source']
    ds0 = xr.open_dataset(path, group='MetaData')
    ds1 = xr.open_dataset(path, group='ObsValue').rename({'snow_water_equivalent':'SWE'})
    ds2 = xr.open_dataset(path, group='ObsError').rename({'snow_water_equivalent':'SWE_ERROR'})
    ds3 = xr.open_dataset(path, group='PreQC').rename({'snow_water_equivalent':'SWE_PreQC'})
    ds = xr.merge([ds0, ds1, ds2, ds3])
    tt = ds.datetime.values
    ttt = [datetime.strptime(t, "%Y-%m-%dT%H:%M:%SZ") for t in tt]
    date = ttt[0].date()
    ds = ds.assign_coords(nlocs = ds.nlocs)
    ds = ds.expand_dims('date')
    ds['date']=[date]
    return ds

def preprocess_owp_obs_snowh(ds):
    path = ds.encoding['source']
    ds0 = xr.open_dataset(path, group='MetaData')
    ds1 = xr.open_dataset(path, group='ObsValue').rename({'snow_depth':'snow_depth'})
    ds2 = xr.open_dataset(path, group='ObsError').rename({'snow_depth':'snow_depth_ERROR'})
    ds3 = xr.open_dataset(path, group='PreQC').rename({'snow_depth':'snow_depth_PreQC'})
    ds = xr.merge([ds0, ds1, ds2, ds3])
    
    tt = ds.datetime.values
    ttt = [datetime.strptime(t, "%Y-%m-%dT%H:%M:%SZ") for t in tt]
    date = ttt[0].date()
    ds = ds.assign_coords(nlocs = ds.nlocs)
    ds = ds.expand_dims('date')
    ds['date']=[date]
    return ds
```

## Check/Vizualize/Validate Output files

```python
files_swe = sorted(glob.glob('/glade/work/mazrooei/jedi/ctest_jedi_workflow/site_09253000/SNOTEL_obs/owp_swe_2021-03-0*'))
ds_in_swe = xr.open_mfdataset(files_swe, preprocess=preprocess_owp_obs_swe) # open_mfdataset(files)
display(ds_in_swe)

files_snowh = sorted(glob.glob('/glade/work/mazrooei/jedi/ctest_jedi_workflow/site_09253000/SNOTEL_obs/owp_depth_2021-03-0*'))
ds_in_snowh = xr.open_mfdataset(files_snowh, preprocess=preprocess_owp_obs_snowh) # open_mfdataset(files)
display(ds_in_snowh)
```

```python
#############################################
##  NOW PROCESSING SWE files
#############################################
ff = ds_in_swe.stack(n=('date','nlocs'))
vv = ff.station_id.values
list_sites = vv[~pd.isnull(vv)]
list_sites = np.unique(list_sites)
print('Found',str(len(list_sites)),'unique sites from swe_obs files')

# list_sites

from tqdm import tqdm

list_ds = list()
for s in tqdm(list_sites):
    oo = ds_in_swe.where(ds_in_swe.station_id == s, drop=True)
    oos = oo.stack(z=('nlocs','date'))
    oos = oos.where(oos.datetime.notnull(), drop=True)
    oos = oos.sortby('datetime')
    
    oo_swe = oos.SWE.values
    oo_swe_er = oos.SWE_ERROR.values
    oo_swe_qc = oos.SWE_PreQC.values
    
    oo_datetime = oos.datetime.values
    oo_datetime2 = [datetime.strptime(t, "%Y-%m-%dT%H:%M:%SZ") for t in oo_datetime]
    oo_x = np.round(np.mean(np.unique(oos.longitude.values)), decimals=3)
    oo_y = np.round(np.mean(np.unique(oos.latitude.values)), decimals=3)
    oo_z = np.round(np.mean(np.unique(oos.height.values)), decimals=3)
    
    da1 = xr.DataArray(data=oo_swe, dims=["time"], coords=dict(time=("time", oo_datetime2))).rename('SWE')
    da2 = xr.DataArray(data=oo_swe_er, dims=["time"], coords=dict(time=("time", oo_datetime2))).rename('SWE_ERROR')
    da3 = xr.DataArray(data=oo_swe_qc, dims=["time"], coords=dict(time=("time", oo_datetime2))).rename('SWE_PreQC')
    ds = xr.merge([da1, da2, da3])
    # da = da.groupby('time.date').mean('time')
    # da['date'] = [datetime.combine(t, datetime.min.time()) for t in da.date.values]
    # print(s)
    # print(da)
    # print(oo_x, oo_y)
    ds = ds.expand_dims('station')
    ds = ds.assign_coords({"lon":("station", [oo_x]),
                           "lat":("station", [oo_y]),
                           "hgt":("station", [oo_z]),
                           "station_id":("station", [s])}
                         )
    
    list_ds.append(ds)

out_owp_obs_swe = xr.concat(list_ds, dim='station')
display(out_owp_obs_swe)


#############################################
##  NOW PROCESSING SNOWH files
#############################################
ff = ds_in_snowh.stack(n=('date','nlocs'))
vv = ff.station_id.values
list_sites = vv[~pd.isnull(vv)]
list_sites = np.unique(list_sites)
print('Found',str(len(list_sites)),'unique sites from depth_obs files')

# list_sites

from tqdm import tqdm

list_ds = list()
for s in tqdm(list_sites):
    oo = ds_in_snowh.where(ds_in_snowh.station_id == s, drop=True)
    oos = oo.stack(z=('nlocs','date'))
    oos = oos.where(oos.datetime.notnull(), drop=True)
    oos = oos.sortby('datetime')
    oo_snow = oos.snow_depth.values
    oo_snow_er = oos.snow_depth_ERROR.values
    oo_snow_qc = oos.snow_depth_PreQC.values
    
    oo_datetime = oos.datetime.values
    oo_datetime2 = [datetime.strptime(t, "%Y-%m-%dT%H:%M:%SZ") for t in oo_datetime]
    
    oo_x = np.round(np.mean(np.unique(oos.longitude.values)), decimals=3)
    oo_y = np.round(np.mean(np.unique(oos.latitude.values)), decimals=3)
    oo_z = np.round(np.mean(np.unique(oos.height.values)), decimals=3)
    
    da1 = xr.DataArray(data=oo_snow, dims=["time"], coords=dict(time=("time", oo_datetime2))).rename('SNOWH')
    da2 = xr.DataArray(data=oo_snow_er, dims=["time"], coords=dict(time=("time", oo_datetime2))).rename('SNOWH_ERROR')
    da3 = xr.DataArray(data=oo_snow_qc, dims=["time"], coords=dict(time=("time", oo_datetime2))).rename('SNOWH_PreQC')
    ds = xr.merge([da1, da2, da3])
    
    # da = da.groupby('time.date').mean('time')
    # da['date'] = [datetime.combine(t, datetime.min.time()) for t in da.date.values]
    # print(s)
    # print(da)
    # print(oo_x, oo_y)
    ds = ds.expand_dims('station')
    ds = ds.assign_coords({"lon":("station", [oo_x]),
                           "lat":("station", [oo_y]),
                           "hgt":("station", [oo_z]),
                           "station_id":("station", [s])}
                         )    
    list_ds.append(ds)

out_owp_obs_snowh = xr.concat(list_ds, dim='station')
display(out_owp_obs_snowh)
```

```python

fig = plt.figure(figsize=(8,4))
for s in out_owp_obs_swe.station.values:
    obs_tmp = out_owp_obs_swe.sel(station=s)
    obs_tmp.SWE.plot.line(ls='none', marker='^', markersize='3', label=obs_tmp.station_id.values, alpha=0.86) #, yscale='log')
plt.ylim([250,800])
plt.title('SNOTEL observations over Little Snake River Near Slater, CO')
plt.legend(loc='best', ncol=5) #, bbox_to_anchor=(0.5, 0., 0.5, 0.5))
# plt.ylim(60,120)
plt.show()


fig = plt.figure(figsize=(8,4))
for s in out_owp_obs_snowh.station.values:
    obs_tmp = out_owp_obs_snowh.sel(station=s)
    obs_tmp.SNOWH.plot.line(ls='none', marker='^', markersize='3', label=obs_tmp.station_id.values, alpha=0.86) #, yscale='log')
plt.ylim([0,6])
plt.title('SNOTEL observations over Little Snake River Near Slater, CO')
plt.legend(loc='best', ncol=5) #, bbox_to_anchor=(0.5, 0., 0.5, 0.5))
# plt.ylim(60,120)
plt.show()
```
